# Simple Linear Regression. Minimal example

### Import the relevant libraries

```python
# We must always import the relevant libraries for our problem at hand. NumPy is a must for this example.
import numpy as np

# matplotlib and mpl_toolkits are not necessary. We employ them for the sole purpose of visualizing the results.  
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Generate random input data to train on

```python
observations = 1000

xs = np.random.uniform(low=-10,high=10,size=(observations,1))
zs = np.random.uniform(-10,10,(observations,1))

inputs = np.column_stack((xs,zs))

print (inputs.shape)
```

Create the targets we will aim at

```python
noise = np.random.uniform(-1,1,(observations,1))

targets = 13*xs + 7*zs -12 + noise

print (targets.shape)
```

Plot the training data

The point is to see that there is a strong trend that our model should learn to reproduce

```python
# In order to use 3D plot, objects should have a shape, so we reshape the targets.
# The method used is reshape and takes the dimensions in which we want to fit the object as arguments
targets = targets.reshape(observations,)

# Plotting according to the matplotlib.pyplot syntax

# Declare the figure
fig = plt.figure()

# Method allowing us to create the 3D plot
ax = fig.add_subplot(111, projection='3d')

# Choose the axis and plis the labels
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')

# The azim parameter can be fiddled with to view the data from different angles. Just change the value of azim=100.
ax.view_init(azim=100)

# Show the plot which we have initialized
plt.show()
targets = targets.reshape(observations,1)
```








Initialize the variables:

```python
init_range = 0.1

weights = np.random.uniform(-init_range, init_range, size=(2,1))

biases = np.random.uniform(-init_range,init_range, size=1)

print(weights)
print(biases)
```

Set a learning rate

```python
learning_rate = 0.01
```

Train the model:

```python
for i in range(250):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    
    loss = np.sum(np.absolute(deltas)) / 2 / observations
    
    print(f'{i+1}) {loss}')
    
    deltas_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights, biases)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
